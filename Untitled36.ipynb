```python
import docx
import pandas as pd
```

```python
filename = "2021 FINAL SSB ABSTRACT SUMMARY.docx"
```

```python
doc = docx.Document(filename)
fullText = []
for para in doc.paragraphs:
    fullText.append(para.text)
```

```python
fullText
```

```python
department_index = []
for index in range(0,len(fullText)):
    if fullText[index] == "SPEARS SCHOOL OF BUSINESS":
        department_index.append(index)
```

```python
dept_project_dict = {}
for i in range(0,len(department_index)):
    dept_projects = []
    if i != len(department_index) - 1:
        last_index = department_index[i+1]
    else:
        last_index = len(fullText) - 1
        
    for j in range(department_index[i]+3, last_index):
        dept_projects.append(fullText[j])
    
    year = fullText[department_index[i]+1].split(' ')[3]
    dept = fullText[department_index[i]+2]
    dept_project_dict[dept.strip() + "-" + year] = dept_projects
```

```python
for key, value in dept_project_dict.items() :
    for index in range(0,len(value)):
        if dept_project_dict[key][index] == ' ':
            dept_project_dict[key][index] = ''
```

```python
for key, value in dept_project_dict.items() :
    for index in range(0,len(value)-1):
        try:
            if dept_project_dict[key][index] == dept_project_dict[key][index+1]:
                del dept_project_dict[key][index+1]
        except:
                print(key, dept_project_dict[key][index])
```

```python
dept_project_index_dict = {}
for key, value in dept_project_dict.items() :
    project_index = []
    for index in range(0,len(value)):
        if value[index] == '':
            project_index.append(index)
    dept_project_index_dict[key] = project_index
```

```python
dataframe = []
for key, value in dept_project_index_dict.items():
    for i in range(0, len(dept_project_index_dict[key])-1):
        try:
            space_index = dept_project_index_dict[key][i]
        
            department = key.split('-')[0].strip()
            year = key.split('-')[1].strip()
            abstract_title = dept_project_dict[key][space_index+1].strip()
            abstract_description = dept_project_dict[key][space_index+2].strip()
            sponsors = dept_project_dict[key][space_index+3].split(':')[1].strip()
            pipds = dept_project_dict[key][space_index+4].split(':')[1].strip()
            
            dataframe.append([department, year, abstract_title, abstract_description, sponsors, pipds])
        except:
            print(key, space_index)
         
            #print(dept_project_dict[key][j])
        
    
```
